```python
zero_days = (df['TotalSteps'] == 0).sum() # number of days with zero steps
zero_days_percentage = (df['TotalSteps'] == 0).mean() *100 # percentage of days with zero steps

print(f"Number of days with zero steps: {zero_days}")
print(f"Percentage of days with zero steps: {zero_days_percentage:.2f}%")
```

```python
active_days = df[df['TotalSteps'] > 0]
active_days["TotalSteps"].describe()
# Analyze the distribution of daily step counts on active days (non-zero steps)
```

```python
df[["VeryActiveMinutes",
    "FairlyActiveMinutes",
    "LightlyActiveMinutes",
    "SedentaryMinutes"]].describe()
# Analyze the distribution of activity minutes across different intensity levels
```

```python
df["SedentaryRatio"] = df["SedentaryMinutes"] / (
    df["VeryActiveMinutes"]
    + df["FairlyActiveMinutes"]
    + df["LightlyActiveMinutes"]
    + df["SedentaryMinutes"]
)

df["SedentaryRatio"].describe()
# Compute sedentary ratio to measure the proportion of daily time spent sedentary
# This normalizes activity levels across days and participants
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 5))
plt.hist(df["TotalSteps"], bins=50, color='skyblue', edgecolor='black')
plt.xlabel('Total Steps / Day')
plt.ylabel('Number of Days')
plt.title('Distribution of Daily Step Counts')
plt.show()
```
