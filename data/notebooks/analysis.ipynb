```python
# Visualize the distribution of daily step counts
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 5))
plt.hist(df["TotalSteps"], bins=50, color='skyblue', edgecolor='black')
plt.xlabel('Total Steps / Day')
plt.ylabel('Number of Days')
plt.title('Distribution of Daily Step Counts')
plt.show()
```

### Interpretation: Distribution of Daily Step Counts

The histogram shows that most daily step counts fall between approximately 3,000 and 12,000 steps, with a noticeable right-skewed distribution.
A smaller number of days record extremely high step counts, indicating occasional highly active days.
The presence of zero-step days highlights periods of inactivity or non-wear, which contributes to the skewness of the distribution.

```python
daily_steps = df.groupby('ActivityDate')['TotalSteps'].mean()
```

```python
# Plot average daily steps over time
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 5))
plt.plot(daily_steps.index, daily_steps.values)
plt.xlabel('Date')
plt.ylabel('Average Steps per Day')
plt.title('Average Daily Steps Over Time')

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Interpretation: Average Daily Steps Over Time

The line chart shows that average daily step counts remain relatively stable throughout the observed time period, 
generally fluctuating between approximately 7,000 and 8,500 steps per day. 
While there are short-term peaks and dips, there is no strong upward or downward long-term trend, 
suggesting consistent overall activity levels across the month. 
The sharp drop near the end of the period may reflect incomplete data for that day rather than a true decline in activity.

```python
# Calculate average activity minutes by intensity level
activity_means = df[["VeryActiveMinutes","FairlyActiveMinutes","LightlyActiveMinutes","SedentaryMinutes"]].mean()
```

```python
# Visualize average activity minutes by intensity level using a bar chart
import matplotlib.pyplot as plt
plt.figure (figsize=(8, 5))
activity_means.plot(kind= "bar", color=["tomato", "orange", "skyblue", "grey"], edgecolor="black")

plt.title ('Average Activity Minutes by Intensity Level')
plt.ylabel ('Average Minutes / Day')
plt.xlabel ('Activity type')
plt.xticks(rotation=0)
plt.show()
```

### Interpretation: Average Activity Minutes by Intensity Level

The bar chart shows that participants spend the majority of their day in sedentary activity, 
with sedentary minutes far exceeding time spent in any active intensity category. 
Lightly active minutes make up the largest portion of active time, while fairly active and very active minutes are comparatively low. 
This suggests that although users engage in some movement throughout the day, most activity occurs at low intensity, 
and sustained moderate-to-vigorous activity is relatively limited.

```python
# Box plot showing spread, median, and outliers in daily step counts
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 4))

plt.boxplot(
    df["TotalSteps"],
    vert=True,
    patch_artist=True,
    boxprops=dict(facecolor="skyblue", color="black"),
    medianprops=dict(color="black"),
    whiskerprops=dict(color="black"),
    capprops=dict(color="black"),
    flierprops=dict(marker='o', markerfacecolor='grey', alpha=0.6)
)

plt.title("Distribution of Daily Step Counts")
plt.ylabel("Total Steps / Day")
plt.show()
```

### Interpretation: Distribution of Daily Step Counts

The box plot shows that daily step counts are right-skewed, with most days falling within a moderate range of steps 
and a smaller number of days with very high step counts. The median daily steps appear around the mid-range, 
while several high-value outliers indicate occasional days of unusually high activity. 
This suggests that while users generally maintain consistent step levels, 
there are sporadic days of intense activity that significantly exceed typical behavior.


```python
plt.figure(figsize=(8, 5))
plt.scatter(
    df["TotalSteps"],
    df["SedentaryMinutes"],
    c=df["SedentaryMinutes"],
    cmap="viridis",
    alpha=0.6
)

plt.xlabel("Total Steps / Day")
plt.ylabel("Sedentary Minutes / Day")
plt.title("Steps vs Sedentary Time (Color-Coded by Sedentary Minutes)")
plt.colorbar(label="Sedentary Minutes")
plt.show()
```

 ### Interpretation: Steps vs Sedentary Time

The scatter plot shows a clear negative relationship between daily step count and sedentary minutes.
As total steps increase, sedentary time generally decreases.
However, the relationship is not perfectly linear. Even on days with moderate step counts,
sedentary time can remain high, suggesting that short bursts of activity do not fully offset
long periods of inactivity.
This indicates that overall daily movement patterns matter more than step count alone.

```python
# Calculate correlation between total steps and sedentary minutes
df["TotalSteps"].corr(df["SedentaryMinutes"])
```

### Interpretation: Correlation Between Steps and Sedentary Time

The correlation between total daily steps and sedentary minutes is approximately **0.33**, indicating a **moderate negative relationship**.

This means that, in general, days with higher step counts tend to have lower sedentary time. However, the relationship is not strong, suggesting that high step counts do not fully eliminate sedentary behavior. This supports earlier visual findings that short bursts of activity may not offset long periods of inactivity throughout the day.
