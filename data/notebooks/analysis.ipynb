```python
zero_days = (df['TotalSteps'] == 0).sum() # number of days with zero steps
zero_days_percentage = (df['TotalSteps'] == 0).mean() *100 # percentage of days with zero steps

print(f"Number of days with zero steps: {zero_days}")
print(f"Percentage of days with zero steps: {zero_days_percentage:.2f}%")
```

```python
active_days = df[df['TotalSteps'] > 0]
active_days["TotalSteps"].describe()
# Analyze the distribution of daily step counts on active days (non-zero steps)
```

```python
df[["VeryActiveMinutes",
    "FairlyActiveMinutes",
    "LightlyActiveMinutes",
    "SedentaryMinutes"]].describe()
# Analyze the distribution of activity minutes across different intensity levels
```

```python
df["SedentaryRatio"] = df["SedentaryMinutes"] / (
    df["VeryActiveMinutes"]
    + df["FairlyActiveMinutes"]
    + df["LightlyActiveMinutes"]
    + df["SedentaryMinutes"]
)

df["SedentaryRatio"].describe()
# Compute sedentary ratio to measure the proportion of daily time spent sedentary
# This normalizes activity levels across days and participants
```

```python
# Visualize the distribution of daily step counts
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 5))
plt.hist(df["TotalSteps"], bins=50, color='skyblue', edgecolor='black')
plt.xlabel('Total Steps / Day')
plt.ylabel('Number of Days')
plt.title('Distribution of Daily Step Counts')
plt.show()
```

```python
daily_steps = df.groupby('ActivityDate')['TotalSteps'].mean()
```

```python
# Plot average daily steps over time
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 5))
plt.plot(daily_steps.index, daily_steps.values)
plt.xlabel('Date')
plt.ylabel('Average Steps per Day')
plt.title('Average Daily Steps Over Time')

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Calculate average activity minutes by intensity level
activity_means = df[["VeryActiveMinutes","FairlyActiveMinutes","LightlyActiveMinutes","SedentaryMinutes"]].mean()
```

```python
# Visualize average activity minutes by intensity level using a bar chart
import matplotlib.pyplot as plt
plt.figure (figsize=(8, 5))
activity_means.plot(kind= "bar", color=["tomato", "orange", "skyblue", "grey"], edgecolor="black")

plt.title ('Average Activity Minutes by Intensity Level')
plt.ylabel ('Average Minutes / Day')
plt.xlabel ('Activity type')
plt.xticks(rotation=0)
plt.show()
```

```python
# Box plot showing spread, median, and outliers in daily step counts
import matplotlib.pyplot as plt
plt.figure(figsize=(6, 4))
plt.boxplot(df["TotalSteps"], vert = True)

plt.title('Distribution of Daily Step Counts')
plt.ylabel('Total Steps / Day')
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))

plt.scatter(df["TotalSteps"], df["SedentaryMinutes"] , color='grey', alpha=0.5)

plt.xlabel('Total Steps / Day')
plt.ylabel('Sedentary Minutes / Day')
plt.title('Relationship Between Daily Steps and Sedentary Minutes')
plt.show()
```
